```python
%matplotlib inline
import os
import re
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def extract_placeholders(directory):
    sfc_placeholders = set()
    tky_placeholders = set()
    pattern = r'blend_of_(.*?)_and_(.*?)__8k__'

    for filename in os.listdir(directory):
        if filename.endswith(".png"):
            matches = re.findall(pattern, filename)
            if matches:
                city1, city2 = matches[0]
                sfc_placeholders.add(city1)
                tky_placeholders.add(city2)

    return list(sfc_placeholders), list(tky_placeholders)

def create_image_grid(directory, sfc_placeholders, tky_placeholders):
    print("X-axis (TKY) placeholders:", tky_placeholders)
    print("Y-axis (SFC) placeholders:", sfc_placeholders)

    # Increase figsize to give more room for y-axis labels
    fig, axes = plt.subplots(len(sfc_placeholders), len(tky_placeholders), figsize=(20, 20), dpi=100)
    
    # Adjust subplot params
    plt.subplots_adjust(left=0.25, right=0.95, bottom=0.05, top=0.95)

    for i, sfc in enumerate(sfc_placeholders):
        for j, tky in enumerate(tky_placeholders):
            filename = f'street_level_view_of_an_ultra_detailed_blend_of_{sfc}_and_{tky}__8k__highly_detailed__cinematic__intricate_details__epic__masterpiece__wide_ang.png'
            filepath = os.path.join(directory, filename)
            if os.path.exists(filepath):
                img = mpimg.imread(filepath)
                axes[i, j].imshow(img)
            axes[i, j].axis('off')

    # Explicitly setting the y-axis labels and drawing the canvas
    for i, ax in enumerate(axes[:, 0]):
        ax.set_ylabel(sfc_placeholders[i], fontsize=10, labelpad=20)
        ax.yaxis.label.set_color('black')  # Make sure the color is visible

    # Drawing the canvas before showing the plot
    fig.canvas.draw()
    plt.show()

# Specify the directory path
directory_path = './images'  # Adjust this to the correct path on your system

# Extract placeholders
sfc_placeholders, tky_placeholders = extract_placeholders(directory_path)

# Create and show the image grid
create_image_grid(directory_path, sfc_placeholders, tky_placeholders)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1 import ImageGrid

# Sample data
im1 = np.arange(100).reshape((10, 10))
im2 = im1.T
im3 = np.flipud(im1)
im4 = np.fliplr(im2)

fig = plt.figure(figsize=(4., 4.))
grid = ImageGrid(fig, 111,  # similar to subplot(111)
                 nrows_ncols=(2, 2),  # creates 2x2 grid of axes
                 axes_pad=0.1,  # pad between axes in inch.
                 )

for ax, im in zip(grid, [im1, im2, im3, im4]):
    # Iterating over the grid returns the Axes.
    ax.imshow(im)

# Set the x-axis and y-axis labels
x_labels = ['X1', 'X2']
y_labels = ['Y1', 'Y2']

# Assuming the order of axes in grid is row-wise from top-left
for i, ax in enumerate(grid):
    if i in [0, 2]:  # First and third plot are on the left
        ax.set_ylabel(y_labels[i // 2])
    if i in [2, 3]:  # Third and fourth plot are on the bottom
        ax.set_xlabel(x_labels[i % 2])

plt.show()
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# Directory where images are stored
directory_path = 'images'
image_files = [os.path.join(directory_path, f) for f in os.listdir(directory_path) if f.endswith('.png')]

# Create the output widget
output = widgets.Output()

# Set up the slider
slider = widgets.IntSlider(value=0, min=0, max=len(image_files)-1, step=1, description='Image Index')

# Slider interaction
def on_value_change(change):
    with output:
        clear_output(wait=True)  # Clear the previous image
        display_image(image_files[slider.value])  # Display the current image

slider.observe(on_value_change, names='value')

# Initial display
with output:
    display_image(image_files[0])  # Show the first image

# Layout
display(slider)
display(output)

# Function to display an image
def display_image(file_path):
    img = mpimg.imread(file_path)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```
